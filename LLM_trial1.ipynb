```python
from transformers import pipeline, set_seed

#Load Model
generator = pipeline("text-generation",
                     model="distilgpt2",
                     pad_token_id=50256)
set_seed(42)

#define chatbot response function
def get_response(user_input):
  #Add a chatbot-like prompt
  prompt = f"User:{user_input}\nBot:"
  response = generator(prompt,max_new_tokens = 50,
                       do_sample=True,
                       temperature = 0.7,
                       top_p=0.95)
  full_text = response[0]["generated_text"]
  #Return just the part after "Bot:"
  if "Bot:" in full_text:
    return full_text.split("Bot:")[-1].strip()
  else:
    return full_text.strip()
```

```python
print(get_response("Hello, How are you today?"))
```

```python
with open("app.py","w") as f:
  f.write("""
  import streamlit as st
  from transformers import pipeline, set_seed

  #Load Model
  generator = pipeline("text-generation", model = "distilgpt2",pad_token = 50256)
  set_seed(42)

  #Define response function
  def get_response(user_input):
    prompt = f"User:{user_input}\\nBot:"
    response = generator(prompt,max_new_tokens=50,do_sample=True, temperature=0.7,top_p=0.95)
    full_text = response[0]["generated_text"]
    if "Bot:" in full_text:
      return full_text.split("Bot:")[-1].strip()
    else:
      return full_text.strip()

  #Streamlit UI
  st.title("LLM_trial1 : Chatbot")
  user_input = st.text_input("You:","")

  if st.button("Send"):
    if user_input:
      response = get_response(user_input)
      st.text_area("Bot:",value=response,height=100)
      """)
```
